```python
zero_cols = [col for col in titanic.columns if 'zero' in col]
titanic = titanic.drop(columns = zero_cols)
```

```python
titanic = titanic.rename(columns={'2urvived': 'Survived'})
```

```python
import seaborn as sns

sns.kdeplot(data=titanic[titanic['Survived']==0], x='Age', label='Not Survived')
sns.kdeplot(data=titanic[titanic['Survived']==1], x='Age', label='Survived')

import matplotlib.pyplot as plt
plt.legend()
plt.show()
```

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived')
```
