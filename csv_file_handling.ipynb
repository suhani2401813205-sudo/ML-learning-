```python
import pandas as pd
```

```python
df = pd.read_csv('train_and_test2.csv')
```

```python
df.head()
```

```python
print("Shape",df.shape)
```

```python
print("\nColumns:",df.columns.tolist())
```

```python
print(df.dtypes)
```

```python
zero_cols =[col for col in df.columns if 'zero' in col]
df = df.drop(columns = zero_cols)
```

```python
print("Remaining columns:",df.columns.tolist())
```

```python
print("New shape :",df.shape)
```

```python
df = df.rename(colums={'2urvived':'Survived'})
```

```python
df = df.rename(columns={'2urvived': 'Survived'})

print(df.columns.tolist())
```

```python
df.info
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df['Embarked'].value_counts())
```

```python
# Most common value se fill karo
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Check karo — ab 0 hona chahiye
print(df.isnull().sum())
```

```python
survivors = df[df['Survived']==1]
print("survived:",len(survivors))
```

```python
# Sirf wo passengers jo 1st class mein the
first_class = df[df['Pclass'] == 1]
print("1st Class passengers:", len(first_class))
```

```python
# Multiple conditions — 1st class AND survived
first_class_survived = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
print("1st class survivors:", len(first_class_survived))
```

```python
# Age 18 se kam wale passengers
children = df[df['Age'] < 18]
print("Children:", len(children))
```

```python
# Sabse zyada fare wale passengers
df.sort_values('Fare', ascending=False).head(10)
```

```python
df.sort_values('Age',ascending=True).head(10)
```

```python
# Multiple columns se sort — Pclass phir Age
df.sort_values(['Pclass', 'Age'], ascending=[True, False]).head(10)
```

```python
# Confirm karne ke liye
print(df.groupby('Sex')['Survived'].mean())
print(df.groupby('Sex')['Survived'].count())
```

```python
# Pclass survival rate
print("Class wise survival:")
print(df.groupby('Pclass')['Survived'].mean())

print("\nGender wise survival:")
print(df.groupby('Sex')['Survived'].mean())

print("\nClass wise average fare:")
print(df.groupby('Pclass')['Fare'].mean())
```

```python
# Fare category banana
df['Fare_Category'] = pd.cut(df['Fare'],
                              bins=[0, 10, 50, 100, 600],
                              labels=['Low', 'Medium', 'High', 'Very High'])

df[['Fare', 'Fare_Category']].head(10)
```

```python
# Age category banana
df['Age_Group'] = pd.cut(df['Age'],
                          bins=[0, 18, 35, 60, 100],
                          labels=['Child', 'Young', 'Middle', 'Senior'])

df[['Age', 'Age_Group']].head(10)
```

```python
import matplotlib.pyplot as plt

# Bar chart — Class wise survival
df.groupby('Pclass')['Survived'].mean().plot(kind='bar',
                                              color=['red','orange','green'],
                                              title='Survival Rate by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

```python
# Histogram — Age distribution
df['Age'].plot(kind='hist',
               bins=20,
               color='steelblue',
               title='Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
# Pie chart — Survived vs Died
df['Survived'].value_counts().plot(kind='pie',
                                    labels=['Died','Survived'],
                                    colors=['red','green'],
                                    autopct='%1.1f%%',
                                    title='Survived vs Died')
plt.show()
```

```python
# Histogram — Age distribution
df['Age'].plot(kind='hist',
               bins=5,
               color='steelblue',
               title='Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
# Fark dekho
df['Age'].plot(kind='hist', bins=5, title='bins=5')
plt.show()

df['Age'].plot(kind='hist', bins=50, title='bins=50')
plt.show()
```

```python
df['Survived'].value_counts().plot(kind='pie',
                                    labels=['Died','Survived'],
                                    colors=['red','green'],
                                    autopct='%1.0f%%')  # No decimal
plt.show()
```

```python
# Age vs Fare ka relation
plt.scatter(df['Age'], df['Fare'],
            c=df['Survived'],  # color = survived ya nahi
            cmap='RdYlGn',     # Red=died, Green=survived
            alpha=0.5)         # transparency

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare (Green=Survived, Red=Died)')
plt.colorbar()
plt.show()
```

```python
# Saare columns ka ek doosre se relation
df.corr(numeric_only=True)
```

```python
# Heatmap — visually dekhna
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,      # numbers dikhao
            cmap='coolwarm', # color scheme
            fmt='.2f')       # 2 decimal places
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Clean dataset save karo
df.to_csv('titanic_clean.csv', index=False)
print("File saved!")
```
